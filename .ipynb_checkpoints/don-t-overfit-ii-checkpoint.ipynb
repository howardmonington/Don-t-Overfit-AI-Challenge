# Don't Overfit II Kaggle Challenge
This is a challenge where the training set has only 250 rows and 302 features whereas the test set has 19,750 rows. In other words, there are 79 times more samples in the test data than in the training data.

The goal of machine learning is to generalize patterns over a dataset. But with such a small dataset, there are more models that can fit the data and there is a higher risk of the model picking up on patterns that don't actually exist. The challenge here is to avoid overfitting to the training data.

# Performing Feature Engineering within the Folds of Stratified Shuffle Split
It is important to oversample within the folds of the cross validation split and not before it. This is because if I oversampled the minority class before cross validation then there is a chance that some of the exact same oversampled data shows up in the training set and the validation set. Oversampling only on the training set also helps to get a better understanding of how the model performs on the real data, without any modifications.

![oversample](oversample_before_cv.jpg)
Link to photo: <a href="https://www.marcoaltini.com/blog/dealing-with-imbalanced-data-undersampling-oversampling-and-proper-cross-validation">source</a>

```python
X = train.drop(['id','target'], axis=1)
y = train['target']
X_test = test.drop(['id'], axis=1)
```

## Scaling
In order to handle the outliers, I am going to use StandardScaler. I am then going to scale the validation dataset within the cross validation folds according to the StandardScaler learned on the training set. This way the test dataset is scaled according to training dataset. I scale the data first, before oversampling so that the oversampling doesn't affect the distribution of the data and therefore impact the scaler.

## Oversampling
In this case, since there are so few training rows, I don't think undersampling would be a good idea since I don't want to reduce the available data even more. I will be using synthetic minority oversampling (SMOTE).  SMOTE creates new (artificial) training examples based on the original training examples. For instance, if it sees two examples (of the same class) near each other, it creates a third artificial one, bang in the middle of the original two. This is different from RandomOverSampler, which simply increases the size of the training data set through repetition of the original examples and does not cause any increase in the variety of training examples.

```python
def classifier_roc_test(n_splits, input_df, input_y, classifier):
    sss = StratifiedShuffleSplit(n_splits = n_splits, test_size = 0.2, random_state = 0)
    roc_scores = []
    for train_index, test_index in sss.split(input_df, input_y):
        X_train, X_test = input_df.loc[train_index,:], input_df.loc[test_index,:]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        # scale the data before oversampling
        cols = input_df.columns
        scaler = StandardScaler()

        X_train = scaler.fit_transform(X_train)
        X_train = pd.DataFrame(X_train, columns = cols)

        X_test = scaler.transform(X_test)
        X_test = pd.DataFrame(X_test, columns = cols)
        
        # create a dataset using SMOTE
        smote = SMOTE(sampling_strategy='minority', n_jobs=-1)
        X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
        
        #outputs X_sm and y_sm as ndarrays, need to convert back to df
        X_resampled = pd.DataFrame(X_resampled, columns=X_train.columns)
        y_resampled = pd.DataFrame(y_resampled, columns=['target'])
        

        classifier.fit(X_resampled, y_resampled)
        y_pred = pd.Series(classifier.predict(X_test)).astype('int64')
        y_test = pd.Series(y_test).astype('int64')
        roc = roc_auc_score(y_test, y_pred)
        roc_scores.append(roc)
        

    return round(statistics.mean(roc_scores), 4)
```

### Avoiding overfitting
Now I'm going to begin trying out some models. The key here will be to avoid complex models with a lot hyperparameters, thus limiting their generalization and possibility of overfitting. I'll set a few of the hyperparameters to avoid overfitting, such as by limiting the tree depth and number of estimators in a random forest model which would limit the model's ability to see patterns and non-existent relationships.

```python
rfc = RandomForestClassifier(n_estimators = 30, max_depth =3, random_state = 0)
etc = ExtraTreesClassifier(n_estimators = 30, max_depth = 3, random_state = 0)
xgb = XGBClassifier(n_estimators = 30, max_depth = 3, random_state =0)
knn =  KNeighborsClassifier(n_neighbors = 5)
sgd = SGDClassifier(random_state = 0, max_iter=50)

models = {'rfc':rfc,'xgb':xgb,'knn':knn,'etc':etc,'sgd':sgd}
```

Based upon the results of this experiment, the sgd classifier seems to be performing the best. I'll move on with using this classifier and perform some feature engineering.

```python
for classifier in models.keys():
    model = models[classifier]
    roc_score = classifier_roc_test(5, X, y, model)
    print(classifier + "  roc: " + str(roc_score))
```

# Splitting data into 1 fold for further testing

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0, shuffle=True)
```

```python
# fit the scaler to the training data, then use that scaler to transform the cv data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_train = pd.DataFrame(X_train, columns = X.columns)

X_val = scaler.transform(X_val)
X_val = pd.DataFrame(X_val, columns = X.columns)
```

```python
# create a dataset using SMOTE
smote = SMOTE(sampling_strategy='minority', n_jobs=-1)
X_train, y_train = smote.fit_resample(X_train, y_train)
        
#outputs X_sm and y_sm as ndarrays, need to convert back to df
X_train = pd.DataFrame(X_train, columns=labels)
y_train = pd.DataFrame(y_train, columns=['target'])
```

## Permutation Importance
From here I can see that although there are 300 columns, not all of them actually come out to be important to the model's predictions. This means that there are a lot of unnecessary columns which I can remove from the data.

Permutation importance randomly re-orders a single column at a time and checks to see how much the accuracy suffers by shuffling that column in order to see which columns were most important in making the predictions.

```python
sgd.fit(X_train, y_train)
perm = PermutationImportance(sgd, random_state=0).fit(X_val, y_val)
eli5.show_weights(perm, feature_names = X_val.columns.tolist(), top = 100)
```

# Univariate Feature Selection
While removing outliers consists of deleting rows from the dataset, feature selection consists of removing columns that do not contribute to the prediction. For this, I'm going to use SelectKBest. There are multiple strategies that I could use. The F-value measures the linear dependency between the feature variable and the target. This means the score might underestimate the relation between a feature and the target if the relationship is nonlinear. The mutual information score is nonparametric and so can capture nonlinear relationships. 

I'm going to use the F-value. The reason why is because the sgd model was trained using the default 'hinge' loss function, which gives a linear SVM. Since it is a linear SVM, I want to select features using a measure of linear dependency.

I need to make sure to only perform the statistical tests on the training set, and then check the accuracy on the validation set. That way I can avoid data leakage.

```python
selector = SelectKBest(f_classif, k=60)
X_new = selector.fit_transform(X_train, y_train)
```

```python
X_new
```

```python
selected_features = pd.DataFrame(selector.inverse_transform(X_new), 
                                 index = X_train.index, columns = labels)
```

```python
selected_features.head()
```

```python
selected_columns = selected_features.columns[selected_features.var() != 0]
```

## Checking to see if the univariate feature selection improved the accuracy
To do this, I'm going to check the accuracy just on the validation set, which was not used at any point to determine the features to keep, thus avoiding overfitting. Here, I can see that it did increase the performance on the validation set. 

```python
roc_score= classifier_roc_test(5, X_val, y_val, sgd)
roc_score
```

```python
roc_score = classifier_roc_test(5, X_val[selected_columns], y_val, sgd)
roc_score
```

Because I thought it would be interesting, I also checked the accuracy using the full dataset. This score is not a reliable though. This is because it has data leakage, since the training data which was used to decide on the features is included when making predictions. So it makes sense that the score would be a lot higher.

```python
roc_score = classifier_roc_test(5, X[selected_columns], y, sgd)
roc_score
```

# Final Model Creation
Now that I know the columns to keep. My final step will be to train the model on the full dataset and make predictions

```python
# scale the full dataset
scaler = StandardScaler()
X_scaled = X[selected_columns]
X_scaled = scaler.fit_transform(X_scaled)

# scale the test dataset according to the same scaler
X_test = test[selected_columns]
X_test = scaler.transform(X_test)
```

```python
# oversample the whole training set using SMOTE
smote = SMOTE(sampling_strategy='minority', n_jobs=-1)
X_scaled, y_scaled = smote.fit_resample(X_scaled, y)
```

```python
# fit the model
sgd.fit(X_scaled, y_scaled)
```
